```
x=[45,34,20,56,84]
plt.pie(x)
plt.show()
```

```
x=[45,34,20,56,84]
lan=['R','JS','Python','c','c++'] #labels names
e=[0.1,0,0.2,0,0] #lokking differnt
plt.pie(x,labels=lan, explode=e,colors=['red','blue','green','yellow','pink'], autopct="%1.1f%%",shadow=True,counterclock=False,
       startangle=90,radius=2,rotatelabels=True) #autopct for percentile, counterclock clockwise or anti
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
n=np.random.randn(1000)
fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot(2,2,1) #rows,columns and no
ax1.hist(n)
ax2=fig.add_subplot(2,2,4) #rows,columns and no
x=[45,34,10,54]
lan=['R','JS','Python','C++'] #labels names
e=[0,0,0,0.2] #lokking differnt
ax2.pie(x,labels=lan, explode=e,colors=['red','green','yellow','pink'], autopct="%1.1f%%",shadow=True,counterclock=False,
       startangle=90,radius=1,rotatelabels=True) #autopct for percentile, counterclock clockwise or anti
plt.show()
```

#SUB PLOT

```
import random
from matplotlib import style
style.use("fivethirtyeight")
fig=plt.figure(figsize=(10,10))
ax1=plt.subplot2grid((6,2),(0,0),rowspan=1,colspan=2)
ax2=plt.subplot2grid((6,2),(1,1),rowspan=4,colspan=1)
ax3=plt.subplot2grid((6,2),(5,0),rowspan=1,colspan=2)

def Create_plots():
  xs=[]
  ys=[]
  for i in range(10):
    x=i
    y=random.randrange(10)
    xs.append(x)
    ys.append(y)
  return xs,ys
  

  

x,y=Create_plots()
ax1.plot(x,y)

x,y=Create_plots()
ax2.plot(x,y)

x,y=Create_plots()
ax3.plot(x,y)
plt.show()
```

```
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import style


#import pandas as pd
#from google.colab import files
#uploaded = files.upload()
#import io
#df = open(io.BytesIO(uploaded['testpy.txt']))

style.use("fivethirtyeight")
fig=plt.figure()
ax1=fig.add_subplot(1,1,1)

def Animate(i):
  graph_data=open("./testpy.txt","r").read()
  lines= graph_data.split("\n")
  xs=[]
  ys=[]
  for line in lines:
    if len(line)>1:
      x,y=line.split(",")
      xs.append(float(x))
      ys.append(float(y))
    ax1.clear()
    ax1.plot(xs.ys)
    
  ani=animation.FuncAnimation(fig,Animate,Interval=1000)
  plt.show()
    
```
