```python
x = [10,20,30,40,50]
```

```python
y = [100,200,300,400,500]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,c = 'red')
plt.show()
```

```python
plt.scatter(x,y, c = ['red','black','blue','green','yellow'])
plt.xlabel('Value of X')
plt.ylabel('Valur of Y')
plt.title('My first scatter plot')

plt.show()
```

```python
import numpy as np
```

```python
np.pi
```

```python
x = np.linspace(-np.pi, np.pi, 200, endpoint=True)
```

```python
len(x)
```

```python
s = np.sin(x)
ss = np.cos(x)
color = np.random.rand(200)
size = 50*np.random.rand(200)
```

```python
plt.plot(x,-s, label = 'Sin', c = 'black')
plt.scatter(x,-ss, label = 'Cos', c = color, s = size)
plt.grid()
plt.legend(loc = 'upper right')
plt.show
```

```python
n= np.random.rand(1000)
plt.hist(n)
plt.show
```

```python
n= np.random.rand(10)
plt.hist(n, bins = 5, color= 'red') # bins are used to divide histogram in specific parts
plt.show
```

```python
x = [45,34,20,56]

language = ['R','JS','C++','Python'] #labels
explode_values = [0.05,0,0,0.1] # values for explode of pi
color = ['red','blue','black','green']
plt.pie(x, labels = language, explode = explode_values, autopct = '%1.1f%%', colors = color, shadow = True, counterclock = False, startangle = 90, radius = 3, rotatelabels = True)# ' autopct '-->format for making percentage sign
plt.show()
```
