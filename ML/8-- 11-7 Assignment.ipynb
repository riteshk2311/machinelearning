```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./housing.csv')
df.head()
```

```python
df.info()
```

```python
df['ocean_proximity'].unique()
```

```python
x10 = df['total_rooms'].loc[df['ocean_proximity']=='NEAR BAY']
x10 = x10.sum()
x20 = df['total_rooms'].loc[df['ocean_proximity']=='<1H OCEAN']
x20 = x20.sum()
x30 = df['total_rooms'].loc[df['ocean_proximity']=='INLAND']
x30 = x30.sum()
x40 = df['total_rooms'].loc[df['ocean_proximity']=='NEAR OCEAN']
x40 = x40.sum()
x50 = df['total_rooms'].loc[df['ocean_proximity']=='ISLAND']
x50 = x50.sum()

X0 = [x10,x20,x30,x40,x50]
label = ['NEAR BAY', '<1H OCEAN', 'INLAND', 'NEAR OCEAN', 'ISLAND']
plt.pie(X0 , labels = label,  autopct = '%1.1f%%', shadow = True,radius = 2, counterclock = False, startangle = 90, rotatelabels = True)
plt.title('ocean_proximity housing data')
plt.show()

```
