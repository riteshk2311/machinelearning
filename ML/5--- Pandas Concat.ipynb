<h1>Pandas Concat</h1>

```python
import pandas as pd
```

```python
indian_weather = pd.DataFrame({
    'City' : ['Mumbai','Delhi','Bhopal'],
    'Temp': [32,45,30],
    'Humidity' : [80, 60, 77]
})

us_weather = pd.DataFrame({
    'City' : ['New York','Chicago','Orlando'],
    'Temp' : [30,47,26,],
    'Humidity' : [56,77,87]
})

pak_weather = pd.DataFrame({
    'City' : ['Lahore','Karanchi','Istanbul'],
    'Temp' : [30,47,26,],
    'Humidity' : [56,77,87],
    'Wind_speed':[200,300,400]
})
```

```python
indian_weather
```

```python
us_weather
```

```python
df = pd.concat([indian_weather,us_weather], ignore_index = True) #concatenation of two data frame
df
```

```python
df = pd.concat([indian_weather,us_weather], keys=['India','Us']) #concatenation of two data frame
df
```

```python
df.loc['India']
```

```python
df.loc['Us']
```

```python
df = pd.concat([indian_weather,us_weather,pak_weather], keys=['India','Us','Pak']) #concatenation of two data frame
df
```
