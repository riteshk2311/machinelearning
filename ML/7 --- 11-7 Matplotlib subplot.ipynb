```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fig = plt.figure(figsize=(10,10))
n= np.random.rand(10)
ax1 = fig.add_subplot(2,2,1) # ROWS COLUMN POSITION
ax1.hist(n)

ax2 = fig.add_subplot(2,2,4) # ROWS COLUMN POSITION
x = [45,34,20,56]

language = ['R','JS','C++','Python'] #labels
explode_values = [0.05,0,0,0.1] # values for explode of pi
color = ['red','blue','black','green']
plt.pie(x, labels = language, explode = explode_values, 
        autopct = '%1.1f%%', colors = color, shadow = True, counterclock = False,
        startangle = 90, rotatelabels = True)# ' autopct '-->format for making percentage sign
plt.show()
```

```python
import random
from matplotlib import style # This import a default style in the matplot lib
style.use('fivethirtyeight') # this is calling of style function... 

fig = plt.figure(figsize = (10,10)) # Back-end running... DONT DELETE THIS

##########################################################
#making subplots......
ax1 = plt.subplot2grid((6,2),(0,0), rowspan=1, colspan=2)
ax2 = plt.subplot2grid((6,2),(1,1), rowspan=4, colspan=1)
ax3 = plt.subplot2grid((6,2),(5,0), rowspan=1, colspan=2)
##########################################################

##########################################################
# making function to generate random number
def Create_plots():
    xs = []
    ys = []
    for i in range(10):
        x = i
        y = random.randrange(10)
        xs.append(x)
        ys.append(y)
    return xs , ys
#########################################################

x, y = Create_plots() # calling for random number
ax1.plot(x,y)
x, y = Create_plots() # calling for random number
ax2.plot(x,y)
x, y = Create_plots() # calling for random number
ax3.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import style

style.use("fivethirtyeight")
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

def Animate(i):
    graph_data = open('./test.txt', 'r').read()
    lines = graph_data.split('\n')
    xs = []
    ys = []
    for line in lines:
        if len(line)>1:
            x, y = line.split(',')
            xs.append(float(x))
            ys.append(float(y))
    ax1.clear()
    ax1.plot(xs, ys)
    
ani = animation.FuncAnimation(fig , Animate, interval = 1000)
plt.show()
```
