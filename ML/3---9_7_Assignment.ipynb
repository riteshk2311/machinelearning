

<h1> <b>Q. Make the graph of temperature and humidity from the 'nyc_weather.csv' file...</b> </h1> 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('nyc_weather.csv')
```

```python
df
```

```python
df.info()
```

```python
temp = df['Temperature']
humidity = df['Humidity']
```

```python
color = np.random.rand(31)
```

```python
plt.scatter(temp,humidity, c = color )
plt.xlabel('Temperature----->')
plt.ylabel('Humidity------->')
plt.title("Temp. & Humidity Graph")
plt.grid()
plt.show
```
